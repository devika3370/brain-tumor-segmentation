```python
imgpath=r'C:\Users\devik\Documents\Academics\Semester 1\Applied ML\Project\BraTS2020_training_data\content\data\volume_340_slice_69.h5'
f=h5py.File(imgpath,'r')
img=f['image']
imgArray=np.array(img)
norm=cv2.normalize(imgArray[:,:,0],None, norm_type=cv2.NORM_MINMAX)
fig = plt.figure(figsize=(22, 6))
plt.imshow(norm, cmap='gray')
plt.title("T1")
```

```python
image_data = f['image'][:]
mask_data = f['mask'][:]
print("Image Data Shape:", image_data.shape)
print("Mask Data Shape:", mask_data.shape)
print(type(image_data))
print(type(mask_data))
print("Data Types - Image:", image_data.dtype, "Mask:", mask_data.dtype)
```

```python
unique_classes, class_counts = np.unique(mask_data, return_counts=True)
print("Unique Classes:", unique_classes)
print("Class Counts:", class_counts)
```

```python
image_mean = np.mean(image_data)
image_std = np.std(image_data)
print("Mean Pixel Value:", image_mean)
print("Standard Deviation:", image_std)
```
